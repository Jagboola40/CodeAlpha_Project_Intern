# Unemployment Rate #

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset #

```python
df= pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
```

```python
df.head(10)
```

```python
# Display information about the dataset
df.info()
```

```python
# Descriptive statitics of the dataset
df.describe()
```

```python
# Group  data by 'Region' and calculate the average unemployment rate for each group
average_unemployment_by_region = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()

average_unemployment_by_region
```

```python
# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df[' Date'])
# Group data by year and calculate the average unemployment rate
average_unemployment_by_year = df.groupby(df['Date'].dt.year)[' Estimated Unemployment Rate (%)'].mean()

# Group data by month and calculate the average unemployment rate
average_unemployment_by_month = df.groupby(df['Date'].dt.month)[' Estimated Unemployment Rate (%)'].mean()



print("\nAverage Unemployment Rate by Year:\n", average_unemployment_by_year)
print("\nAverage Unemployment Rate by Month:\n", average_unemployment_by_month)
```

```python
# Plotting average unemployment rate by year
plt.figure(figsize=(10, 6))
plt.bar(average_unemployment_by_year.index, average_unemployment_by_year.values, color='skyblue')
plt.xlabel("Year")
plt.ylabel("Average Unemployment Rate (%)")
plt.title("Average Unemployment Rate by Year")
plt.xticks(average_unemployment_by_year.index) # Ensure all years are shown on x-axis
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Plotting average unemployment rate by month
plt.figure(figsize=(10, 6))
plt.plot(average_unemployment_by_month.index, average_unemployment_by_month.values, marker='o', linestyle='-')
plt.xlabel("Month")
plt.ylabel("Average Unemployment Rate (%)")
plt.title("Average Unemployment Rate by Month")
plt.xticks(average_unemployment_by_month.index) # Ensure all months are shown on x-axis
plt.grid(axis='y', linestyle='--')
plt.show()
```
